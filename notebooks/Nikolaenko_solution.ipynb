## 3. Предобработка данных

### 3.1 Обработка аномалий, кодирование и создание признаков

**Аномалии:**
- `Current Loan Amount = 99999999` → заменяем на NaN, но сохраняем флаг `Is Bad Loan Amount` — сам факт отсутствия данных о сумме несёт информацию
- `Credit Score > 900` → делим на 10 (например, 7260 → 726)

**Флаги пропусков** (создаются до заполнения медианой):
- `Annual_Income_Missing`, `Credit_Score_Missing` — ~20% пропусков; люди без данных о доходе или скоре могут принадлежать к другой группе риска

**Кодирование категорий:**
- `Years in current job` — порядковый маппинг (< 1 year → 0 … 10+ years → 10)
- `Term` — бинарный (Long Term → 1)
- `Home Ownership`, `Purpose` — LabelEncoder; при обучении LightGBM передаются как нативные категории через `categorical_feature`, что позволяет модели находить правильные сплиты без ложного порядка

**Заполнение пропусков:** медианой по обучающей выборке; `Months since last delinquent` → 0 (нет просрочки)

**Созданные признаки:**

| Признак | Смысл |
|---|---|
| `Has Delinquent` | Была ли вообще просрочка (1/0) |
| `Debt to Income Ratio` | Годовой долг / годовой доход |
| `Credit Utilization` | Текущий баланс / максимальный открытый кредит |
| `Loan to Income Ratio` | Сумма кредита / годовой доход |
| `Risk Score` | Сумма: кредитные проблемы + банкротства + налоговые залоги |
| `Monthly Debt to Balance` | Ежемесячный долг / текущий кредитный баланс |
| `Accounts per Credit Year` | Число открытых счетов на год кредитной истории |
| `Score x Credit Problems` | Взаимодействие: высокий скор при наличии проблем — подозрительный паттерн |
| `Bankruptcy x Monthly Debt` | Взаимодействие: банкрот с высоким ежемесячным долгом — повышенный риск |

````python
job_years_map = {
    '< 1 year': 0, '1 year': 1, '2 years': 2, '3 years': 3, '4 years': 4,
    '5 years': 5, '6 years': 6, '7 years': 7, '8 years': 8, '9 years': 9, '10+ years': 10
}

CAT_COLS = ['Home Ownership', 'Purpose']
LOG_COLS = ['Annual Income', 'Monthly Debt', 'Current Credit Balance', 'Maximum Open Credit']

def preprocess(df, fit_encoders=None):
    df = df.copy()

    df['Is Bad Loan Amount'] = (df['Current Loan Amount'] == 99999999).astype(int)
    df['Current Loan Amount'] = df['Current Loan Amount'].replace(99999999, np.nan)
    df['Credit Score'] = df['Credit Score'].apply(
        lambda x: x / 10 if pd.notna(x) and x > 900 else x
    )

    df['Annual_Income_Missing'] = df['Annual Income'].isna().astype(int)
    df['Credit_Score_Missing']  = df['Credit Score'].isna().astype(int)

    df['Has Delinquent'] = df['Months since last delinquent'].notna().astype(int)
    annual_income_safe = df['Annual Income'].replace(0, np.nan)
    max_credit_safe    = df['Maximum Open Credit'].replace(0, np.nan)
    df['Debt to Income Ratio'] = (df['Monthly Debt'] * 12) / annual_income_safe
    df['Credit Utilization']   = df['Current Credit Balance'] / max_credit_safe
    df['Loan to Income Ratio'] = df['Current Loan Amount'] / annual_income_safe
    df['Risk Score'] = (df['Number of Credit Problems']
                        + df['Bankruptcies'].fillna(0)
                        + df['Tax Liens'])

    df['Years in current job'] = df['Years in current job'].map(job_years_map)
    df['Term'] = (df['Term'] == 'Long Term').astype(int)

    encoders = fit_encoders if fit_encoders is not None else {}
    for col in CAT_COLS:
        if fit_encoders is None:
            le = LabelEncoder()
            df[col] = le.fit_transform(df[col].astype(str))
            encoders[col] = le
        else:
            le = encoders[col]
            known = set(le.classes_)
            df[col] = df[col].apply(lambda x: le.transform([x])[0] if x in known else -1)

    df['Months since last delinquent'] = df['Months since last delinquent'].fillna(0)

    fill_cols = ['Annual Income', 'Years in current job', 'Credit Score',
                 'Current Loan Amount', 'Bankruptcies', 'Tax Liens',
                 'Debt to Income Ratio', 'Credit Utilization',
                 'Loan to Income Ratio', 'Risk Score']
    medians = encoders.get('medians', {})
    for col in fill_cols:
        if col not in df.columns:
            continue
        if fit_encoders is None:
            medians[col] = df[col].median()
        df[col] = df[col].fillna(medians.get(col, df[col].median()))

    # Ratio и interaction признаки
    df['Monthly Debt to Balance']   = df['Monthly Debt'] / df['Current Credit Balance'].replace(0, np.nan)
    df['Accounts per Credit Year']  = df['Number of Open Accounts'] / df['Years of Credit History'].replace(0, np.nan)
    df['Score x Credit Problems']   = df['Credit Score'] * df['Number of Credit Problems']
    df['Bankruptcy x Monthly Debt'] = df['Bankruptcies'] * df['Monthly Debt']

    # Log-трансформы для правосторонних распределений
    for col in LOG_COLS:
        df[f'Log {col}'] = np.log1p(df[col].clip(lower=0))

    # Бинарные флаги
    df['Has Credit Problems'] = (df['Number of Credit Problems'] > 0).astype(int)
    df['Has Bankruptcies']    = (df['Bankruptcies'] > 0).astype(int)

    # Новые признаки: обобщённый флаг риска, взаимодействие возраст×скор, свежесть просрочки
    df['High Risk'] = (
        (df['Number of Credit Problems'] > 0) |
        (df['Bankruptcies'] > 0) |
        (df['Tax Liens'] > 0)
    ).astype(int)
    df['Credit Age x Score']  = df['Years of Credit History'] * df['Credit Score']
    df['Delinquent Recency']  = df['Has Delinquent'] / (df['Months since last delinquent'] + 1)

    df['Income per Account']  = df['Annual Income'] / df['Number of Open Accounts'].replace(0, np.nan)
    df['Credit Score Norm']   = df['Credit Score'] / df['Years of Credit History'].replace(0, np.nan)

    extra_cols = [
        'Monthly Debt to Balance', 'Accounts per Credit Year',
        'Score x Credit Problems', 'Bankruptcy x Monthly Debt',
        'Income per Account', 'Credit Score Norm',
        'High Risk', 'Credit Age x Score', 'Delinquent Recency',
    ] + [f'Log {c}' for c in LOG_COLS]
    for col in extra_cols:
        if fit_encoders is None:
            medians[col] = df[col].median()
        df[col] = df[col].fillna(medians.get(col, df[col].median()))

    if fit_encoders is None:
        encoders['medians'] = medians
        return df, encoders
    return df
````

````python
df_train_proc, encoders = preprocess(df_train)
df_test_proc = preprocess(df_test, fit_encoders=encoders)

print('Пропуски в train после обработки:', df_train_proc.isnull().sum().sum())
print('Пропуски в test после обработки: ', df_test_proc.isnull().sum().sum())
print('Строк в тесте (должно быть 2500):', len(df_test_proc))
df_train_proc.head()
````

### 3.2 Target Encoding для категориальных признаков

Заменяем метку категории (например, `Own Home = 2`) на **долю дефолтов в этой группе** — это несёт прямую информацию о риске. Применяем OOF-схему (out-of-fold), чтобы избежать утечки таргета в обучающую выборку.

````python
def add_target_encoding(df_tr_proc, df_te_proc, df_tr_orig, df_te_orig,
                        cat_cols=None, n_splits=5, smooth=10):
    """OOF target encoding: заменяем категорию на долю дефолтов в группе (с smoothing)."""
    if cat_cols is None:
        cat_cols = ['Home Ownership', 'Purpose']
    df_tr_proc = df_tr_proc.copy()
    df_te_proc = df_te_proc.copy()
    y_tr = df_tr_orig['Credit Default']
    global_mean = y_tr.mean()
    kf = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=SEED)

    for col in cat_cols:
        oof = np.zeros(len(df_tr_orig))
        for tr_idx, val_idx in kf.split(df_tr_orig, y_tr):
            stats = (df_tr_orig.iloc[tr_idx]
                     .groupby(col)['Credit Default']
                     .agg(['sum', 'count']))
            sm = (stats['sum'] + global_mean * smooth) / (stats['count'] + smooth)
            oof[val_idx] = df_tr_orig.iloc[val_idx][col].map(sm).fillna(global_mean).values
        df_tr_proc[f'{col}_TE'] = oof

        # Для теста: статистики по полному train
        stats_full = df_tr_orig.groupby(col)['Credit Default'].agg(['sum', 'count'])
        sm_full = (stats_full['sum'] + global_mean * smooth) / (stats_full['count'] + smooth)
        df_te_proc[f'{col}_TE'] = df_te_orig[col].map(sm_full).fillna(global_mean).values

    return df_tr_proc, df_te_proc

df_train_proc, df_test_proc = add_target_encoding(
    df_train_proc, df_test_proc, df_train, df_test
)
te_cols = [c for c in df_train_proc.columns if c.endswith('_TE')]
print(f'Target Encoding добавлен: {te_cols}')
print('\nСредняя доля дефолтов по Home Ownership:')
print(df_train[['Home Ownership', 'Credit Default']]
      .groupby('Home Ownership')['Credit Default'].mean().round(3))
print('\nTE-значения (должны совпадать с долями выше):')
print(df_train_proc.groupby('Home Ownership_TE')['Home Ownership_TE']
      .count().head())
````

````python
# Ранговые признаки: percentile rank клиента внутри обучающей выборки.
# Rank(x) = доля объектов, у которых значение <= x.
# Это убирает эффект выбросов и даёт модели «контекст» — насколько клиент хуже/лучше среднего.
RANK_COLS = ['Credit Score', 'Annual Income', 'Monthly Debt',
             'Debt to Income Ratio', 'Current Credit Balance']

for col in RANK_COLS:
    if col in df_train_proc.columns:
        df_train_proc[f'{col}_Rank'] = df_train_proc[col].rank(pct=True)
        # Для теста: ранг относительно обучающего распределения (без утечки)
        train_vals = df_train_proc[col].values
        df_test_proc[f'{col}_Rank'] = df_test_proc[col].apply(
            lambda x: float((train_vals <= x).mean())
        )

rank_cols_added = [f'{c}_Rank' for c in RANK_COLS if c in df_train_proc.columns]
print(f'Добавлено ранговых признаков: {len(rank_cols_added)}')
print(df_train_proc[rank_cols_added].describe().round(3))
````

## 5. Базовая модель — Logistic Regression

````python
X_train, X_val, y_train, y_val = train_test_split(
    X, y, test_size=0.2, random_state=SEED, stratify=y
)
print(f'Train: {X_train.shape}, Val: {X_val.shape}')
print(f'Баланс классов — 0: {(y_train==0).sum()}, 1: {(y_train==1).sum()}')
````

````python
def evaluate_model(model, X_tr, y_tr, X_v, y_v, name='', fit_params=None):
    model.fit(X_tr, y_tr, **(fit_params or {}))
    y_pred = model.predict(X_v)
    f1   = f1_score(y_v, y_pred)
    prec = precision_score(y_v, y_pred)
    rec  = recall_score(y_v, y_pred)
    print(f'{name:30s}  F1={f1:.4f}  Precision={prec:.4f}  Recall={rec:.4f}')
    return f1, prec, rec

lr = LogisticRegression(max_iter=1000, class_weight='balanced', random_state=SEED)
evaluate_model(lr, X_train, y_train, X_val, y_val, 'Logistic Regression')
````

**Вывод:** Логистическая регрессия — базовая интерпретируемая модель. Ограничена линейными границами решений, поэтому нелинейные зависимости улавливает хуже.

## 6. Сравнение моделей

````python
pos_weight = (y_train == 0).sum() / (y_train == 1).sum()

models = {
    'Logistic Regression': (
        LogisticRegression(max_iter=1000, class_weight='balanced', random_state=SEED),
        {}
    ),
    'Random Forest': (
        RandomForestClassifier(n_estimators=200, class_weight='balanced',
                               random_state=SEED, n_jobs=-1),
        {}
    ),
    'LightGBM': (
        lgb.LGBMClassifier(n_estimators=300, scale_pos_weight=pos_weight,
                           random_state=SEED, n_jobs=-1, verbose=-1),
        {'categorical_feature': CAT_COLS}
    ),
    'XGBoost': (
        XGBClassifier(n_estimators=300, scale_pos_weight=pos_weight,
                      random_state=SEED, n_jobs=-1,
                      eval_metric='logloss', verbosity=0),
        {}
    ),
    'CatBoost': (
        CatBoostClassifier(iterations=300, scale_pos_weight=pos_weight,
                           random_seed=SEED, verbose=0),
        {}
    ),
}

results = []
for name, (model, fp) in models.items():
    f1, prec, rec = evaluate_model(model, X_train, y_train, X_val, y_val, name,
                                   fit_params=fp)
    results.append({'Модель': name, 'F1': f1, 'Precision': prec, 'Recall': rec})

results_df = pd.DataFrame(results).set_index('Модель')
print('\n=== Итоговое сравнение ===')
print(results_df.round(4))
````

## 7. Подбор гиперпараметров LightGBM (Optuna)

````python
pos_weight_ratio = (y_train == 0).sum() / (y_train == 1).sum()
print(f'Соотношение классов (0/1): {pos_weight_ratio:.2f}')

def objective(trial):
    params = {
        'num_leaves':        trial.suggest_int('num_leaves', 20, 300),
        'learning_rate':     trial.suggest_float('learning_rate', 0.005, 0.2, log=True),
        'n_estimators':      trial.suggest_int('n_estimators', 200, 1000),
        'min_child_samples': trial.suggest_int('min_child_samples', 5, 150),
        'subsample':         trial.suggest_float('subsample', 0.5, 1.0),
        'colsample_bytree':  trial.suggest_float('colsample_bytree', 0.5, 1.0),
        'reg_alpha':         trial.suggest_float('reg_alpha', 0.0, 2.0),
        'reg_lambda':        trial.suggest_float('reg_lambda', 0.0, 2.0),
        # Диапазон: от 0.5 (акцент на precision) до полного pos_weight (компенсация дисбаланса)
        'scale_pos_weight':  trial.suggest_float('scale_pos_weight', 0.5, pos_weight_ratio),
        'max_depth':         trial.suggest_int('max_depth', 3, 12),
        'min_gain_to_split': trial.suggest_float('min_gain_to_split', 0.0, 1.0),
    }
    cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=SEED)
    f1s, precs, recs = [], [], []
    for tr_idx, val_idx in cv.split(X_train, y_train):
        m = lgb.LGBMClassifier(random_state=SEED, n_jobs=-1, verbose=-1, **params)
        m.fit(X_train.iloc[tr_idx], y_train.iloc[tr_idx], categorical_feature=CAT_COLS)
        pred = m.predict(X_train.iloc[val_idx])
        f1s.append(f1_score(y_train.iloc[val_idx], pred))
        precs.append(precision_score(y_train.iloc[val_idx], pred, zero_division=0))
        recs.append(recall_score(y_train.iloc[val_idx], pred))

    avg_f1 = np.mean(f1s)
    avg_p  = np.mean(precs)
    avg_r  = np.mean(recs)

    # Квадратичный штраф за P < 0.50: precision важнее recall.
    # Линейный штраф за R < 0.45 — не позволяем полностью игнорировать recall.
    balance = min(avg_p / 0.50, 1.0) ** 2 * min(avg_r / 0.45, 1.0)
    return avg_f1 * balance

study = optuna.create_study(
    direction='maximize',
    sampler=optuna.samplers.TPESampler(seed=SEED)
)
study.optimize(objective, n_trials=100, show_progress_bar=True)

print('\nЛучшие гиперпараметры:')
print(study.best_params)
print(f'\nCross-val balanced score: {study.best_value:.4f}')
````

````python
best_model = lgb.LGBMClassifier(
    random_state=SEED, n_jobs=-1, verbose=-1,
    **study.best_params
)
best_model.fit(X_train, y_train, categorical_feature=CAT_COLS)
y_pred_val   = best_model.predict(X_val)
y_pred_train = best_model.predict(X_train)

f1_tr  = f1_score(y_train, y_pred_train)
f1_val = f1_score(y_val,   y_pred_val)
gap    = f1_tr - f1_val

print('=== Проверка переобучения (Train vs Val) ===')
print(f'  Train F1  = {f1_tr:.4f}')
print(f'  Val   F1  = {f1_val:.4f}')
print(f'  Gap       = {gap:.4f}  {"[!] возможно переобучение" if gap > 0.08 else "[ok] норма"}')
print()
print('=== Финальная модель (LightGBM + Optuna, порог 0.5) ===')
print(f'F1        = {f1_val:.4f}')
print(f'Precision = {precision_score(y_val, y_pred_val):.4f}')
print(f'Recall    = {recall_score(y_val, y_pred_val):.4f}')
print()
print(classification_report(y_val, y_pred_val, target_names=['Нет дефолта (0)', 'Дефолт (1)']))
````

**Вывод:** Optuna с алгоритмом TPE (Tree-structured Parzen Estimator) и 100 итерациями подобрал оптимальные гиперпараметры LightGBM. TPE обучается на предыдущих результатах и концентрирует поиск в перспективных областях. Пространство поиска: `num_leaves`, `max_depth`, `learning_rate`, `n_estimators`, `min_child_samples`, `subsample`, `colsample_bytree`, `reg_alpha`, `reg_lambda`, `min_gain_to_split`, `scale_pos_weight`. Целевая функция использует **квадратичный штраф по Precision** (`(P/0.50)²`) и линейный штраф по Recall (`R/0.45`), что направляет поиск в сторону P ≥ 0.50. Категориальные признаки передавались через `categorical_feature` на каждом fold.

## 7.5 Подбор гиперпараметров CatBoost и XGBoost (Optuna) + стекинг

Для объединения трёх бустингов (LightGBM, CatBoost, XGBoost) используем **стекинг**: мета-модель (LogisticRegression с `class_weight='balanced'`) обучается на OOF-вероятностях трёх базовых моделей. Мета-модель сама определяет оптимальные веса, тогда как простое усреднение игнорирует различные сильные стороны каждой модели. Гиперпараметры CatBoost и XGBoost подбираются Optuna (50 итераций каждая) с той же квадратичной целевой функцией по Precision.

````python
def objective_cb(trial):
    params = {
        'iterations':          trial.suggest_int('iterations', 200, 800),
        'learning_rate':       trial.suggest_float('learning_rate', 0.01, 0.2, log=True),
        'depth':               trial.suggest_int('depth', 3, 8),
        'l2_leaf_reg':         trial.suggest_float('l2_leaf_reg', 0.5, 10.0),
        'subsample':           trial.suggest_float('subsample', 0.6, 1.0),
        'scale_pos_weight':    trial.suggest_float('scale_pos_weight', 0.5, pos_weight_ratio),
        'min_data_in_leaf':    trial.suggest_int('min_data_in_leaf', 5, 100),
    }
    cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=SEED)
    f1s, precs, recs = [], [], []
    for tr_idx, val_idx in cv.split(X_train, y_train):
        m = CatBoostClassifier(**params, random_seed=SEED, verbose=0)
        m.fit(X_train.iloc[tr_idx], y_train.iloc[tr_idx])
        pred = m.predict(X_train.iloc[val_idx])
        f1s.append(f1_score(y_train.iloc[val_idx], pred))
        precs.append(precision_score(y_train.iloc[val_idx], pred, zero_division=0))
        recs.append(recall_score(y_train.iloc[val_idx], pred))
    avg_f1 = np.mean(f1s)
    avg_p  = np.mean(precs)
    avg_r  = np.mean(recs)
    balance = min(avg_p / 0.50, 1.0) ** 2 * min(avg_r / 0.45, 1.0)
    return avg_f1 * balance

study_cb = optuna.create_study(direction='maximize',
                                sampler=optuna.samplers.TPESampler(seed=SEED))
study_cb.optimize(objective_cb, n_trials=50, show_progress_bar=True)
print('Лучшие параметры CatBoost:', study_cb.best_params)
print(f'CV balanced score: {study_cb.best_value:.4f}')

# --- XGBoost с балансированной целевой функцией ---
def objective_xgb(trial):
    params = {
        'n_estimators':      trial.suggest_int('n_estimators', 200, 800),
        'learning_rate':     trial.suggest_float('learning_rate', 0.01, 0.2, log=True),
        'max_depth':         trial.suggest_int('max_depth', 3, 8),
        'subsample':         trial.suggest_float('subsample', 0.6, 1.0),
        'colsample_bytree':  trial.suggest_float('colsample_bytree', 0.5, 1.0),
        'reg_alpha':         trial.suggest_float('reg_alpha', 0.0, 2.0),
        'reg_lambda':        trial.suggest_float('reg_lambda', 0.0, 2.0),
        'scale_pos_weight':  trial.suggest_float('scale_pos_weight', 0.5, pos_weight_ratio),
        'min_child_weight':  trial.suggest_int('min_child_weight', 1, 20),
    }
    cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=SEED)
    f1s, precs, recs = [], [], []
    for tr_idx, val_idx in cv.split(X_train, y_train):
        m = XGBClassifier(**params, random_state=SEED, n_jobs=-1,
                          eval_metric='logloss', verbosity=0)
        m.fit(X_train.iloc[tr_idx], y_train.iloc[tr_idx])
        pred = m.predict(X_train.iloc[val_idx])
        f1s.append(f1_score(y_train.iloc[val_idx], pred))
        precs.append(precision_score(y_train.iloc[val_idx], pred, zero_division=0))
        recs.append(recall_score(y_train.iloc[val_idx], pred))
    avg_f1 = np.mean(f1s)
    avg_p  = np.mean(precs)
    avg_r  = np.mean(recs)
    balance = min(avg_p / 0.50, 1.0) ** 2 * min(avg_r / 0.45, 1.0)
    return avg_f1 * balance

study_xgb = optuna.create_study(direction='maximize',
                                  sampler=optuna.samplers.TPESampler(seed=SEED))
study_xgb.optimize(objective_xgb, n_trials=50, show_progress_bar=True)
print('Лучшие параметры XGBoost:', study_xgb.best_params)
print(f'CV balanced score: {study_xgb.best_value:.4f}')

best_catboost = CatBoostClassifier(**study_cb.best_params, random_seed=SEED, verbose=0)
best_catboost.fit(X_train, y_train)

best_xgboost = XGBClassifier(**study_xgb.best_params, random_state=SEED, n_jobs=-1,
                               eval_metric='logloss', verbosity=0)
best_xgboost.fit(X_train, y_train)

# === Стекинг 3 моделей: LGBM + CatBoost + XGBoost ===
# class_weight='balanced' в мета-модели — компенсирует дисбаланс классов
# в OOF-предсказаниях и делает вероятности более откалиброванными.
print('\n--- Стекинг (3 модели): OOF-вероятности ---')
cv5 = StratifiedKFold(n_splits=5, shuffle=True, random_state=SEED)
oof_lgbm = np.zeros(len(X_train))
oof_cb   = np.zeros(len(X_train))
oof_xgb  = np.zeros(len(X_train))

for fold_i, (tr_idx, val_idx) in enumerate(cv5.split(X_train, y_train), 1):
    m_l = lgb.LGBMClassifier(**study.best_params, random_state=SEED, n_jobs=-1, verbose=-1)
    m_l.fit(X_train.iloc[tr_idx], y_train.iloc[tr_idx], categorical_feature=CAT_COLS)
    oof_lgbm[val_idx] = m_l.predict_proba(X_train.iloc[val_idx])[:, 1]

    m_c = CatBoostClassifier(**study_cb.best_params, random_seed=SEED, verbose=0)
    m_c.fit(X_train.iloc[tr_idx], y_train.iloc[tr_idx])
    oof_cb[val_idx] = m_c.predict_proba(X_train.iloc[val_idx])[:, 1]

    m_x = XGBClassifier(**study_xgb.best_params, random_state=SEED, n_jobs=-1,
                         eval_metric='logloss', verbosity=0)
    m_x.fit(X_train.iloc[tr_idx], y_train.iloc[tr_idx])
    oof_xgb[val_idx] = m_x.predict_proba(X_train.iloc[val_idx])[:, 1]
    print(f'  fold {fold_i}/5 done')

meta_lr = LogisticRegression(C=1.0, random_state=SEED, class_weight='balanced')
meta_lr.fit(np.column_stack([oof_lgbm, oof_cb, oof_xgb]), y_train)
w = meta_lr.coef_[0]
print(f'Веса: LGBM={w[0]:.3f}  CB={w[1]:.3f}  XGB={w[2]:.3f}')

# Оценка на валидации
proba_lgbm_val = best_model.predict_proba(X_val)[:, 1]
proba_cb_val   = best_catboost.predict_proba(X_val)[:, 1]
proba_xgb_val  = best_xgboost.predict_proba(X_val)[:, 1]
meta_X_val     = np.column_stack([proba_lgbm_val, proba_cb_val, proba_xgb_val])
proba_stacked_val = meta_lr.predict_proba(meta_X_val)[:, 1]
proba_ens_val     = proba_stacked_val

print('\n=== Сравнение при пороге 0.5 ===')
for name, proba in [('LightGBM', proba_lgbm_val),
                    ('CatBoost', proba_cb_val),
                    ('XGBoost',  proba_xgb_val),
                    ('Стекинг',  proba_stacked_val)]:
    pred = (proba >= 0.5).astype(int)
    print(f'{name:12s}  F1={f1_score(y_val, pred):.4f}  '
          f'P={precision_score(y_val, pred):.4f}  '
          f'R={recall_score(y_val, pred):.4f}')
````

## 8. Оптимизация порога классификации

По умолчанию модель использует порог 0.5. При дисбалансе классов оптимальный порог может быть другим. Перебираем пороги от 0.10 до 0.90 с шагом 0.005 и выбираем порог с **максимальным F1 при условии P ≥ 0.50 и R ≥ 0.45** — это гарантирует, что Precision не упадёт ниже целевого уровня. Если такого порога нет — fallback на глобальный argmax F1.

````python
from sklearn.metrics import precision_score as prec_score, recall_score as rec_score

y_proba_val = proba_ens_val  # вероятности стекинга

thresholds  = np.arange(0.10, 0.90, 0.005)
f1_by_thr   = np.array([f1_score(y_val, (y_proba_val >= t).astype(int))
                         for t in thresholds])
prec_by_thr = np.array([prec_score(y_val, (y_proba_val >= t).astype(int), zero_division=0)
                         for t in thresholds])
rec_by_thr  = np.array([rec_score(y_val, (y_proba_val >= t).astype(int))
                         for t in thresholds])

# Ищем порог с максимальным F1 при P >= 0.50 и R >= 0.45
# При P=R=0.50 F1=0.50 — это минимум; требуем хотя бы одно из условий жёстко
mask = (prec_by_thr >= 0.50) & (rec_by_thr >= 0.45)
if mask.any():
    best_idx  = np.argmax(np.where(mask, f1_by_thr, 0.0))
    strategy  = 'max F1 при P >= 0.50 и R >= 0.45'
else:
    best_idx  = np.argmax(f1_by_thr)
    strategy  = 'max F1 (fallback — граница P/R не достигнута)'

best_threshold = thresholds[best_idx]
best_f1_thr    = f1_by_thr[best_idx]
p_at_best      = prec_by_thr[best_idx]
r_at_best      = rec_by_thr[best_idx]

print(f'Стандартный порог 0.500:')
y_pred_05 = (y_proba_val >= 0.5).astype(int)
print(f'  F1={f1_score(y_val, y_pred_05):.4f}  '
      f'Precision={prec_score(y_val, y_pred_05):.4f}  '
      f'Recall={rec_score(y_val, y_pred_05):.4f}')
print(f'\nОптимальный порог {best_threshold:.3f} ({strategy}):')
print(f'  F1={best_f1_thr:.4f}  Precision={p_at_best:.4f}  Recall={r_at_best:.4f}')

fig, ax = plt.subplots(figsize=(11, 4))
ax.plot(thresholds, f1_by_thr,   color='steelblue', linewidth=2, label='F1-score')
ax.plot(thresholds, prec_by_thr, color='tomato',    linewidth=2, linestyle='--', label='Precision')
ax.plot(thresholds, rec_by_thr,  color='seagreen',  linewidth=2, linestyle='--', label='Recall')
ax.axvline(best_threshold, color='black', linestyle='--',
           label=f'Оптимум: {best_threshold:.3f}  (F1={best_f1_thr:.3f}, P={p_at_best:.3f}, R={r_at_best:.3f})')
ax.axvline(0.5, color='gray', linestyle=':', label='Стандартный: 0.500')
ax.axhline(0.50, color='orange', linestyle=':', linewidth=1.5, label='Минимум 0.50')
ax.set_xlabel('Порог классификации')
ax.set_ylabel('Значение метрики')
ax.set_title('Стекинг: F1, Precision, Recall по порогам')
ax.legend()
plt.tight_layout()
plt.show()

y_pred_opt = (y_proba_val >= best_threshold).astype(int)
print('\n=== Метрики стекинга с оптимальным порогом ===')
print(classification_report(y_val, y_pred_opt,
                             target_names=['Нет дефолта (0)', 'Дефолт (1)']))

fig, ax = plt.subplots(figsize=(6, 5))
ConfusionMatrixDisplay.from_predictions(
    y_val, y_pred_opt,
    display_labels=['Нет дефолта', 'Дефолт'],
    cmap='Blues', ax=ax
)
ax.set_title(f'Матрица ошибок — стекинг, порог {best_threshold:.3f}', fontsize=12)
plt.tight_layout()
plt.show()
````

## 9. Прогнозы для тестового датасета

````python
# Обучаем все три модели на ПОЛНОМ train
final_lgbm = lgb.LGBMClassifier(random_state=SEED, n_jobs=-1, verbose=-1,
                                  **study.best_params)
final_lgbm.fit(X, y, categorical_feature=CAT_COLS)

final_cb = CatBoostClassifier(**study_cb.best_params, random_seed=SEED, verbose=0)
final_cb.fit(X, y)

final_xgb = XGBClassifier(**study_xgb.best_params, random_state=SEED, n_jobs=-1,
                            eval_metric='logloss', verbosity=0)
final_xgb.fit(X, y)

# OOF на полном X → финальная мета-модель
print('Вычисляем OOF для финальной мета-модели (полный train)...')
cv5_full = StratifiedKFold(n_splits=5, shuffle=True, random_state=SEED)
oof_l_full = np.zeros(len(X))
oof_c_full = np.zeros(len(X))
oof_x_full = np.zeros(len(X))

for fold_i, (tr_idx, val_idx) in enumerate(cv5_full.split(X, y), 1):
    m_l = lgb.LGBMClassifier(**study.best_params, random_state=SEED, n_jobs=-1, verbose=-1)
    m_l.fit(X.iloc[tr_idx], y.iloc[tr_idx], categorical_feature=CAT_COLS)
    oof_l_full[val_idx] = m_l.predict_proba(X.iloc[val_idx])[:, 1]

    m_c = CatBoostClassifier(**study_cb.best_params, random_seed=SEED, verbose=0)
    m_c.fit(X.iloc[tr_idx], y.iloc[tr_idx])
    oof_c_full[val_idx] = m_c.predict_proba(X.iloc[val_idx])[:, 1]

    m_x = XGBClassifier(**study_xgb.best_params, random_state=SEED, n_jobs=-1,
                         eval_metric='logloss', verbosity=0)
    m_x.fit(X.iloc[tr_idx], y.iloc[tr_idx])
    oof_x_full[val_idx] = m_x.predict_proba(X.iloc[val_idx])[:, 1]
    print(f'  fold {fold_i}/5 done')

final_meta_lr = LogisticRegression(C=1.0, random_state=SEED, class_weight='balanced')
final_meta_lr.fit(np.column_stack([oof_l_full, oof_c_full, oof_x_full]), y)

X_test = df_test_proc[FEATURES]
meta_X_test = np.column_stack([final_lgbm.predict_proba(X_test)[:, 1],
                                final_cb.predict_proba(X_test)[:, 1],
                                final_xgb.predict_proba(X_test)[:, 1]])
proba_stacked_test = final_meta_lr.predict_proba(meta_X_test)[:, 1]
y_test_pred        = (proba_stacked_test >= best_threshold).astype(int)

print(f'\nПорог классификации: {best_threshold:.3f}')
print('Форма тестовой выборки:', X_test.shape)
print('Класс 0:', (y_test_pred == 0).sum(), '| Класс 1:', (y_test_pred == 1).sum())
````

### 2. Предобработка данных
- `Current Loan Amount = 99999999` → NaN; `Credit Score > 900` → делим на 10
- Создан флаг **`Is Bad Loan Amount`** — сохраняем информацию об аномальной строке даже после замены на NaN
- Созданы **флаги пропусков** `Annual_Income_Missing` и `Credit_Score_Missing` до заполнения медианой: клиенты без данных о доходе или скоре могут принадлежать к другой группе риска
- `Months since last delinquent` → 0 (нет просрочки); остальные пропуски → медиана обучающей выборки
- `Years in current job` — порядковый маппинг; `Term` — бинарный; `Home Ownership`, `Purpose` — LabelEncoder

---

### 3. Инженерия признаков
Итоговый набор содержит **28 признаков** (16 исходных + 12 созданных):

| Признак | Тип | Обоснование |

### 4. Сравнение базовых моделей
Сравнивались 5 моделей на валидационной выборке (80/20 стратифицированный split):
- **Logistic Regression** — базовая линейная модель (`class_weight='balanced'`)
- **Random Forest** — ансамбль деревьев (`class_weight='balanced'`)
- **LightGBM** — градиентный бустинг с нативными категориями и `scale_pos_weight`
- **XGBoost** — градиентный бустинг с `scale_pos_weight`
- **CatBoost** — градиентный бустинг с `scale_pos_weight`

---

### 5. Подбор гиперпараметров (Optuna, алгоритм TPE)
Все три бустинга подобраны Optuna с 5-кратной стратифицированной кросс-валидацией: LightGBM — 100 итераций, CatBoost и XGBoost — по 50.

**Пространство поиска LightGBM:** `num_leaves`, `max_depth`, `learning_rate`, `n_estimators`, `min_child_samples`, `subsample`, `colsample_bytree`, `reg_alpha`, `reg_lambda`, `min_gain_to_split`, `scale_pos_weight`.

**Целевая функция (одинакова для всех трёх моделей):**
```
score = F1 × (min(P / 0.50, 1.0))² × min(R / 0.45, 1.0)
```
Квадратичный штраф по Precision сильнее наказывает за P < 0.50, линейный штраф по Recall не позволяет полностью игнорировать Recall.

---

### 6. Стекинг трёх моделей
LightGBM, CatBoost и XGBoost объединены через **стекинг**: мета-модель (LogisticRegression, `class_weight='balanced'`) обучается на OOF-вероятностях трёх базовых моделей. Это позволяет автоматически взвесить вклад каждой модели в зависимости от её сильных сторон.

---

### 7. Оптимизация порога классификации
Перебирались пороги от 0.10 до 0.90 с шагом 0.005. Выбирается порог с **максимальным F1 при условии P ≥ 0.50 и R ≥ 0.45** (если такой порог существует; иначе — fallback на глобальный argmax F1).
